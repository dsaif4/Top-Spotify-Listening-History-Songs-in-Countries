```python
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import warnings
warnings.filterwarnings("ignore")
from datetime import datetime, timedelta
```

```python
# top 10 artist by total listing time 
top_artist = df.groupby('artist_name')['duration'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,8))
sns.barplot(y = top_artist.index, x = top_artist.values)
plt.title("Top Artist by total time")
plt.xlabel("Total Duration played")
plt.ylabel("Artist name")
plt.show()
```

```python
# top track most played songs
most_played = df.groupby('track_name')['duration'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,8))
sns.barplot(x = most_played.values, y = most_played.index)
plt.title("Top played songs")
plt.xlabel("Total Duration played")
plt.ylabel("Track name")
plt.show()
```

```python
# most played albums bu listing time 
top_album = df.groupby('album_name')['duration'].sum().sort_values(ascending=False).head()
plt.figure(figsize=(10,8))
sns.barplot(y = top_album.index, x = top_album.values)
plt.title("Top 10 Album")
plt.xlabel("Total Duration played")
plt.ylabel("Album name")
plt.show()
```

```python
# Add time-based features
df['date'] = df['ts'].dt.date
df['year'] = df['ts'].dt.year
df['month'] = df['ts'].dt.month
df['day'] = df['ts'].dt.day
df['weekday'] = df['ts'].dt.day_name()
df['hour'] = df['ts'].dt.hour
```

```python
# check the data again
df.head()
```

```python
# Listening trends by hour of day
hourlytrend_eachDay = df.groupby('hour')['duration'].sum()
plt.figure(figsize=(10,7))
sns.lineplot(x = hourlytrend_eachDay.index, y = hourlytrend_eachDay.values, marker = 'o')
plt.title("Hourly song distribution per day")
plt.xlabel("Duration")
plt.ylabel("Total Duration")
plt.xticks(range(0,24))
plt.grid(True)
plt.show()
```

```python
# Listening trends over a week
weekly_trend = df.groupby('weekday')['duration'].sum()
weekdays = ['Sunday','Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
plt.figure(figsize=(10,6))
sns.lineplot(x = weekdays,  y =weekly_trend[weekdays] , marker = 'o')
plt.title("Trend over a Week")
plt.xlabel("Days")
plt.ylabel('Total Duration')
plt.grid(True)
plt.show()
```

```python
# Monthly listening trend over a year
monthly_trend = df.groupby(['month', 'year'])['duration'].sum().reset_index()
monthly_trend['year_month'] = pd.to_datetime(monthly_trend[['year', 'month']].assign(day =1))
plt.figure(figsize=(10,6))
sns.lineplot(data = monthly_trend, x = 'year_month' , y = 'duration' ,marker = 'o')
plt.title("Monthly listening trend over a months")
plt.xlabel("Year")
plt.ylabel("Total Duration")
plt.grid(True)
plt.show()
```

```python
# Sort data by timestamp
df = df.sort_values('ts').reset_index(drop=True)

# Compute time difference between consecutive plays
df['time_diff'] = df['ts'].diff().fillna(pd.Timedelta(seconds=0))

# Define a session break as 30 minutes or more between tracks
session_threshold = pd.Timedelta(minutes=30)
df['new_session'] = df['time_diff'] > session_threshold
df['session_id'] = df['new_session'].cumsum()

# Analyze session stats
session_summary = df.groupby('session_id').agg(
    session_start=('ts', 'min'),
    session_end=('ts', 'max'),
    total_tracks=('track_name', 'count'),
    total_minutes=('duration', 'sum')
).reset_index()

# Add session start hour for temporal patterns
session_summary['start_hour'] = session_summary['session_start'].dt.hour

# Plot session length distribution (in minutes)
plt.figure(figsize=(10,5))
sns.histplot(session_summary['total_minutes'], bins=40, kde=True)
plt.title(' Distribution of Session Durations')
plt.xlabel('Total Minutes per Session')
plt.ylabel('Count')
plt.show()

# Plot session counts by start hour
plt.figure(figsize=(10,5))
sns.countplot(x='start_hour', data=session_summary, palette='pastel')
plt.title(' Number of Sessions by Start Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Sessions')
plt.grid(True)
plt.show()

# Display session summary stats
session_summary.describe()[['total_tracks', 'total_minutes']]
```
